```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn import datasets
```

```python
data = datasets.load_iris(return_X_y=True)
```

```python
X = pd.DataFrame(data=data[0])
```

```python
y = data[1]
```

```python
X.shape
```

```python
X['target'] = y 
```

```python
X.head()
```

```python
y = pd.Series(y)
```

```python
y.shape
```

```python
sns.pairplot(data=X.iloc[:,:],hue='target')
```

```python
fig,axes = plt.subplots(2,2,figsize=(16,6))
i=0
j=0
for ft in X.columns[:-1]:
    sns.histplot(data=X,x=ft,kde='True',bins=20,ax=axes[i][j],)
    if j==1:
        i+=1
        j=-1
    j+=1
```

```python
sns.heatmap(X.corr())
```

```python
X.drop('target',axis = 1,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=4)
```

```python
from naiveBayes import NaiveBayes
naive = NaiveBayes()
```

```python
y_pred = naive.fit_gaussian(X_train,y_train,X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy_score(y_pred=y_pred,y_true=y_test)
```

```python
confusion_matrix(y_test,y_pred)
```
